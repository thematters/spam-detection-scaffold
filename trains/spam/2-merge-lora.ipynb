```python
import os
import torch
from pathlib import Path
from peft import AutoPeftModelForSequenceClassification
from transformers import AutoTokenizer
```

```python
# 替換成 1-train.ipynb 訓練後挑選的最佳 checkpoint 路徑
checkpoint = 'spam/training-run/checkpoint-best'
```

```python
p = Path(checkpoint)
```

```python
postfix = Path(checkpoint).parts[1]
```

```python
output_dir = f'merged_{postfix}/'
```

```python
output_model_tar = f'spam_{postfix}.tar'
```

```python
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
tokenizer.save_pretrained(output_dir)
```

```python
model = AutoPeftModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2, torch_dtype='auto', cache_dir='.cache')
```

```python
model = model.merge_and_unload(progressbar=True, safe_merge=True)
```

```python
model.dtype
```

```python
model.save_pretrained(output_dir)
```
