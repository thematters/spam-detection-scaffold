```python
import pandas as pd
```

```python
%%time
data1 = pd.read_parquet('spam-labels-input.parquet.gzip')
```

```python
%%time
# spam.csv 應有以下欄位：article_id, title, summary, summary_customized, content, is_spam
# 內容收錄已被人工標記 is_spam 的文章，以及來自需要審核之作者的文章。
data2 = pd.read_csv('spam.csv')
```

```python
data1.info()
```

```python
data2.info()
```

```python
data2['summary_customized'] = data2['summary_customized'].apply(lambda x : 1 if x=='t' else 0)
```

```python
data2['is_spam'] = data2['is_spam'].apply(lambda x : 0 if x=='f' else 1)
```

```python
data2['is_spam'].value_counts()
```

```python
data = pd.concat([data1, data2], ignore_index=True)
```

```python
data.info()
```

```python
data['is_spam'].value_counts()
```

```python
%%time
result = data.drop_duplicates(subset=['title', 'content'], keep='last')
```

```python
result.info()
```

```python
result['is_spam'].value_counts()
```

```python
max_id = max(data["article_id"])
```

```python
result.to_parquet('spam-labels-merged.parquet.gzip', compression='gzip')
```
